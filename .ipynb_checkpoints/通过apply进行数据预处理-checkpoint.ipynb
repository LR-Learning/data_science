```python
# 创建一个函数统一处理
def foo(line):
    items = line.strip().split(' ')
    return Series([items[1],items[3],items[5]])
```

```python
df_temp = df['data'].apply(foo)
```

```python
df_temp.head()
```

```python
df_temp.rename(columns={0:"Symbol", 1:"Seqno", 2:"Price"})
```

```python
df_new = df.combine_first(df_temp)
```
