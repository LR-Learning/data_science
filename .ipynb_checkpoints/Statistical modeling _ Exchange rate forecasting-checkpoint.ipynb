```python
%matplotlib inline
import pandas as pd
import pandas_datareader
import datetime
import matplotlib.pylab as plt
import seaborn as sns
from matplotlib.pylab import style
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import statsmodels.api as sm

style.use('ggplot')
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
exchangeFile = 'D:/document/统计建模大赛/数据/CNY_USD.csv'
exchange = pd.read_csv(exchangeFile, index_col=0, parse_dates=[0])
exchange.head(10)
```

```python
exchange_week = exchange['Close'].resample('W-MON').mean() # 以每周的平均值为重采样的值
exchange_train = exchange_week['20100104' : '20210101'] # 选择训练数据范围
exchange_train.tail(10)
```

```python
exchange_train.plot(figsize=(12,8))
plt.legend(bbox_to_anchor=(1.25, 0.5))
plt.title("Exchange Clsoe")
sns.despine()
```

```python
# 对数据进行平稳化操作
exchange_diff = exchange_train.diff()
exchange_diff = exchange_diff.dropna()

plt.figure()
plt.plot(exchange_diff)
plt.title('一阶差分')
plt.show()
```

```python
exchange_diff_diff = exchange_train.diff().diff()
exchange_diff_diff = exchange_diff_diff.dropna()

plt.figure()
plt.plot(exchange_diff_diff)
plt.title('二阶差分')
plt.show()
```

```python
# 用于决定Q的值
acf = plot_acf(exchange_diff, lags=20)
plt.title('ACF')
acf.show()
```

```python
# 用于决定P的值
pacf = plot_pacf(exchange_diff, lags=20)
plt.title('PACF')
pacf.show()
```

```python
exchange_train[np.isnan(exchange_train)] = 6.5
exchange_train[np.isinf(exchange_train)] = 6.5
model = ARIMA(exchange_train, order=(1,1,1), freq='W-MON')
```

```python
result = model.fit()
print(result.summary())
```

```python
pred = result.predict('2019', '20210831', dynamic=True, typ='levels')
print(pred)
```

```python
plt.figure(figsize=(6,6))
plt.xticks(rotation=45)
plt.plot(pred)
plt.plot(exchange_train)
```
