```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
%matplotlib inline
import seaborn as sns
```

```python
plt.hist(s1) # 画一个直方图
```

```python
s1.plot(kind='kde') # 画一个密度图
```

```python
sns.distplot(s1,hist=True,kde=False) # 通过seaborn画图
```

```python
sns.displot(s1)
```

```python
sns.histplot(s1)
```

```python
sns.kdeplot(s1)
```
