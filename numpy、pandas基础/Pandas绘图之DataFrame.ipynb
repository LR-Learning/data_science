```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
df = DataFrame(
    np.random.randint(1,10,40).reshape(10,4),
    columns=['A','B','C','D']
)
```

```python
df
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='barh')
```

```python
df.plot(kind='bar', stacked=True)
```
